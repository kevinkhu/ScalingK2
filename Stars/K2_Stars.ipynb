From [Scaling K2. I. Revised Parameters for 222,088 K2 Stars and a K2 Planet Radius Valley at 1.9 R<sub>âŠ•</sub>](https://ui.adsabs.harvard.edu/abs/2020ApJS..247...28H/abstract).

Import useful libraries.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from astroquery.vizier import Vizier
```

Read in the K2 stellar parameters table from Vizier as an astropy table.

```python
Vizier.ROW_LIMIT=-1
table = Vizier.query_constraints(catalog="J/ApJS/247/28/table1")
k2starcat = table[table.keys()[0]]
k2starcat
```

Convert table to a Pandas DataFrame.

```python
k2stars = k2starcat.to_pandas()
k2stars
```

Keep in mind there are stars that were observed in multiple campaigns. For a table of unique targets:

```python
k2stars = k2stars.drop_duplicates(subset=['EPIC'])
k2stars.reset_index(inplace=True,drop=True)
k2stars
```
